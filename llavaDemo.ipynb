##import necessari

````python
import requests #serve per prendere le immagini online
from PIL import Image # gestione immagini
import torch
from transformers import BitsAndBytesConfig
from transformers import pipeline
from transformers import LlavaProcessor, LlavaForConditionalGeneration
````

##Quantization config

````python
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)
````

##Caricamento del modello

````python
model_id = "llava-hf/llava-1.5-7b-hf"

pipe = pipeline("image-to-text", model=model_id, model_kwargs={"quantization_config": quantization_config})
````

Ricorda che il prompt deve essere in questo formato:
```bash
USER: <image>\n<prompt>\nASSISTANT:
```

##Caricamento immagine

````python
image = Image.open("/content/drive/MyDrive/Immagini test/testGenerated.png")
image
````

##test di esempio

````python
max_new_tokens = 200
prompt = "USER: <image>\nDescribe this pitcure and where does she come from?\nASSISTANT:"
outputs = pipe(image, prompt=prompt, generate_kwargs={"max_new_tokens": 200})
print(outputs[0]["generated_text"])
````
